## Question 1. Make an API request that returns the months where "fire" was searched in 2016. Which month had the most searches? 

```python
endpoint ='https://data.nashville.gov/resource/fuaa-r5cm.json'
```

```python
params = {'year':'2016', 'query_text':'fire' }
```

```python
response = requests.get(endpoint, params=params)
```

```python
response
```

```python
result = response.json()
```

```python
result_df=pd.DataFrame(result)
```

```python
result_df.sort_values(by='query_count',ascending=False)
```
